```python
import ast


def convert_cast(str):  # top 5 cast
    tags = []
    for i in ast.literal_eval(str)[:5]:
        tags.append(i['name'])
    return tags
```

```python
s= str(movies['cast'][0])
```

```python
convert_cast(s)
```

```python
movies.cast= movies.cast.apply(convert_cast)
```

```python
import ast


def convert_crew(str):  # only want the director
    tags = []
    for i in ast.literal_eval(str):
        if i['job']=='Director':
            tags.append(i['name'])
            break
    return tags
```

```python
s= movies.crew[0]
# s
```

```python
convert_crew(s)
```

```python
movies.crew = movies.crew.apply(convert_crew)
```

`stemming`

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps= PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
    
```

```python
s= new_df.tags[0]
s
```

```python
stem(s)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

## `vectorization`

```python
from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features=3500, stop_words='english') # 5000 words
```

```python
import numpy as np
```

```python
vectors= cv.fit_transform(new_df.tags).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import cosine_distances
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index= new_df[new_df.title==movie].index[0]
    distances= similarity[movie_index]
    similar_movies= sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in similar_movies:
        print(i[0])
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Mission: Impossible')
```
